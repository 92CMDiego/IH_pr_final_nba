```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('all_team_rosters_1998_2022.csv')
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['HEIGHT'].value_counts()
```

```python
data[['FEET', 'INCHES']] = data['HEIGHT'].str.split('-', expand=True)

data[['FEET', 'INCHES']] = data[['FEET', 'INCHES']].astype(int)

data['HEIGHT_CM'] = (data['FEET'] * 30.48) + (data['INCHES'] * 2.54)
```

```python
data['HEIGHT_CM'] = round(data['HEIGHT_CM'], 0)
```

```python
data['HEIGHT_CM'] = data['HEIGHT_CM'].astype(int)
```

```python
data['WEIGHT_KG'] = data['WEIGHT'] * 0.45359237
```

```python
data
```

```python
data['WEIGHT_KG'] = round(data['WEIGHT_KG'], 2)
```

```python
# data.to_csv('data.csv', index=False)
```

Seasonal height means.

```python
seasonal_height_means = data.groupby('SEASON')['HEIGHT_CM'].agg('mean')
```

```python
seasonal_height_means
```

Seasonal height means by position.

```python
data['POSITION'].value_counts()
```

```python
seasonal_position_height_means = data.groupby(['SEASON', 'POSITION'])['HEIGHT_CM'].agg('mean')
```

```python
seasonal_position_height_means
```

```python
seasonal_position_height_means_df = pd.DataFrame(seasonal_position_height_means.reset_index())
```

```python
seasonal_position_height_means_df['HEIGHT_CM'] = round(seasonal_position_height_means_df['HEIGHT_CM'], 0)
```

```python
seasonal_position_height_means_df['HEIGHT_CM'] = seasonal_position_height_means_df['HEIGHT_CM'].astype(int)
```

```python
seasonal_position_height_means_df.dtypes
```

```python
# seasonal_position_height_means_df.to_csv('seasonal_position_height_means_1998_2022.csv', index=False)
```

```python
seasonal_position_height_means_df[seasonal_position_height_means_df['POSITION']=='G']
```

Weights.

```python
seasonal_position_weight_means = data.groupby(['SEASON', 'POSITION'])['WEIGHT_KG'].agg('mean')
```

```python
seasonal_position_weight_means
```

```python
seasonal_position_weight_means_df = pd.DataFrame(seasonal_position_weight_means.reset_index())
```

```python
seasonal_position_weight_means_df['WEIGHT_KG'] = round(seasonal_position_weight_means_df['WEIGHT_KG'], 0)
```

```python
seasonal_position_weight_means_df['WEIGHT_KG'] = seasonal_position_weight_means_df['WEIGHT_KG'].astype(int)
```
