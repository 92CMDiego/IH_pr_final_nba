USA / NON_USA flag.

```python
df['USA'] = (df['COUNTRY'] == 'USA').astype(int)
df['NON_USA'] = (df['COUNTRY'] != 'USA').astype(int)
```

Grouping by season and USA / NON_USA.

```python
seasonal_usa_players = df.groupby('SEASON')[['USA']].sum().reset_index()
seasonal_nonusa_players = df.groupby('SEASON')[['NON_USA']].sum().reset_index()
```

```python
print(len(seasonal_usa_players))
len(seasonal_nonusa_players)
```

```python
seasonal_usa_nonusa = seasonal_usa_players.merge(seasonal_nonusa_players, on='SEASON')
```

```python
# seasonal_usa_nonusa
```

```python
seasonal_usa_nonusa['PERC_USA'] = (seasonal_usa_nonusa['USA'] / (seasonal_usa_nonusa['USA'] + seasonal_usa_nonusa['NON_USA'])) * 100
seasonal_usa_nonusa['PERC_NONUSA'] = (seasonal_usa_nonusa['NON_USA'] / (seasonal_usa_nonusa['USA'] + seasonal_usa_nonusa['NON_USA'])) * 100
```

International players per team.

```python
usa_nonusa_season_team = df.groupby(['SEASON', 'TeamID']).agg({'USA': 'sum', 'NON_USA': 'sum'}).reset_index()
```
