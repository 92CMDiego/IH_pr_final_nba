```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

Get a list of all CSV files in the directory.

```python
import glob

csv_files = glob.glob('./mvp_votes_data/*.txt')

data = pd.DataFrame()

year = 1999

for f in csv_files:
    df = pd.read_csv(f)
    df['YEAR'] = year
    year = year + 1
    data = pd.concat([data, df])
```
