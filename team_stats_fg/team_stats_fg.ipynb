```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('all_teams_yby_stats_1949_2022.csv')
```

```python
data.head()
```

```python
data = data[(data['YEAR'] >= '1998-99') & (data['YEAR'] <= '2022-23')]
```

```python
data
```

```python
data.groupby(['YEAR'])['TEAM_ID'].count()
```

```python
data['FGM_PER_GAME'] = round(data['FGM'] / data['GP'], 1)
data['FGA_PER_GAME'] = round(data['FGA'] / data['GP'], 1)
data['FG3M_PER_GAME'] = round(data['FG3M'] / data['GP'], 1)
data['FG3A_PER_GAME'] = round(data['FG3A'] / data['GP'], 1)
data['FTM_PER_GAME'] = round(data['FTM'] / data['GP'], 1)
data['FTA_PER_GAME'] = round(data['FTA'] / data['GP'], 1)
```

```python
data.isna().sum()
```

```python
data
```

```python
# data.dtypes
```

```python
data[data['FTM_PER_GAME']==0]
```

```python
data['FG2M'] = data['FGM'] - data['FG3M']
data['FG2A'] = data['FGA'] - data['FG3A']
```

```python
data['FG2M_PER_GAME'] = data['FG2M'] / data['GP']
data['FG2A_PER_GAME'] = data['FG2A'] / data['GP']
```

```python
data['FG2_PCT'] = data['FG2M'] / data['FG2A']
```

3-pointers.

```python
seasonal_fg3a_mean = data.groupby('YEAR').agg({'FG3A': 'mean'}).round().astype(int).reset_index()
seasonal_fg3m_mean = data.groupby('YEAR').agg({'FG3M': 'mean'}).round().astype(int).reset_index()
```

```python
seasonal_fg3m_mean
```

```python
print(len(seasonal_fg3a_mean))
len(seasonal_fg3m_mean)
```

```python
seasonal_fg3_mean = seasonal_fg3a_mean.merge(seasonal_fg3m_mean, on='YEAR')
```

```python
seasonal_fg3_mean
```
