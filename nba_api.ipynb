### Teams year by year stats.

```python
from nba_api.stats.endpoints import teamyearbyyearstats

all_teams_stats_df = pd.DataFrame()

for id in team_ids:
       
    team_stats = teamyearbyyearstats.TeamYearByYearStats(team_id=id)
    team_stats_df = team_stats.get_data_frames()[0]
    
    all_teams_stats_df = pd.concat([all_teams_stats_df, team_stats_df], ignore_index=True)
```

#### Players' bio stats.

```python
import pandas as pd
from nba_api.stats.endpoints import leaguedashplayerbiostats
from datetime import date

all_player_G_bio_stats_df = pd.DataFrame()

current_year = 2023

for year in range(current_year - 25, current_year):
    
    season = f"{year}-{str(year + 1)[2:]}"
    
    player_bio_stats = leaguedashplayerbiostats.LeagueDashPlayerBioStats(season=season, player_position_abbreviation_nullable='G')
    player_bio_stats_data = player_bio_stats.get_data_frames()[0]
    
    player_bio_stats_data['SEASON'] = season
    
    all_player_G_bio_stats_df = pd.concat([all_player_G_bio_stats_df, player_bio_stats_data], ignore_index=True)
```

### League leaders.

```python
from nba_api.stats.endpoints import leagueleaders
import pandas as pd
from datetime import date

all_league_leaders_df = pd.DataFrame()

current_year = date.today().year

for year in range(current_year - 20, current_year):
    season = f"{year}-{str(year + 1)[2:]}"
    league_leaders = leagueleaders.LeagueLeaders(
            season=season,
            season_type_all_star="Regular Season",
        )

    league_leaders_data = league_leaders.get_data_frames()[0]
    league_leaders_data['SEASON'] = season

    all_league_leaders_df = pd.concat([all_league_leaders_df, league_leaders_data], ignore_index=True)
```

```python
from nba_api.stats.endpoints import leagueleaders
import pandas as pd
from datetime import date

all_league_leaders_df = pd.DataFrame()

current_year = date.today().year

for year in range(current_year - 20, current_year):
    season = f"{year}-{str(year + 1)[2:]}"
    league_assist_leaders = leagueleaders.LeagueLeaders(
        season=season,
        season_type_all_star="Regular Season",
        stat_category_abbreviation='AST'
    )

    league_assist_leaders_data = league_assist_leaders.get_data_frames()[0]
    league_assist_leaders_data['SEASON'] = season

    all_league_leaders_df = pd.concat([all_league_leaders_df, league_assist_leaders_data], ignore_index=True)
```

```python
import pandas as pd
from nba_api.stats.endpoints import teamdashboardbygeneralsplits
import random
import time

current_year = 2023

all_teams_dashboard_df = pd.DataFrame()

for year in range(current_year - 25, current_year):
    season = f"{year}-{str(year + 1)[2:]}"
    for id in team_ids:
        team_dashboard = teamdashboardbygeneralsplits.TeamDashboardByGeneralSplits(team_id=id, season=season, measure_type_detailed_defense='Advanced')
        sleep_time = random.uniform(0.5, 1.0)
        time.sleep(sleep_time)
        team_dashboard_data = team_dashboard.get_data_frames()[0]
        team_dashboard_data['TEAM_ID'] = id
        all_teams_dashboard_df = pd.concat([all_teams_dashboard_df, team_dashboard_data], ignore_index=True)
```

### Team rosters.

```python
import pandas as pd
from nba_api.stats.endpoints import commonteamroster

all_team_rosters = pd.DataFrame()

current_year = 2023

for year in range(current_year - 25, current_year):
    
    season = f"{year}-{str(year + 1)[2:]}"
    
    for id in team_ids:

        team_roster = commonteamroster.CommonTeamRoster(season=season, team_id=id)
        team_roster_data = team_roster.get_data_frames()[0]
            
        all_team_rosters = pd.concat([all_team_rosters, team_roster_data], ignore_index=True)
```
