```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

Players stats.

```python
from nba_api.stats.endpoints import leaguedashplayerstats

player_stats_1998_2022 = pd.DataFrame()
current_year = 2023

for year in range(current_year - 25, current_year):
    
    season = f"{year}-{str(year + 1)[2:]}"

    player_stats = leaguedashplayerstats.LeagueDashPlayerStats(season=season, measure_type_detailed_defense='Base', per_mode_detailed='PerGame', season_type_all_star='Regular Season')
    player_stats_df = player_stats.get_data_frames()[0]
        
    player_stats_df['SEASON'] = season
    
    player_stats_1998_2022 = pd.concat([player_stats_1998_2022, player_stats_df], ignore_index=True)
```

```python
player_stats_1998_2022
```

```python
player_stats_1998_2022.isna().sum().sum()
```

Teams stats.

```python
from nba_api.stats.static import teams

nba_teams = teams.get_teams()
team_ids = [team['id'] for team in nba_teams]
```

```python
from nba_api.stats.endpoints import teamyearbyyearstats

all_teams_stats_df = pd.DataFrame()

for id in team_ids:
    team_stats = teamyearbyyearstats.TeamYearByYearStats(team_id=id, per_mode_simple='PerGame', season_type_all_star='Regular Season')
    team_stats_df = team_stats.get_data_frames()[0]
    all_teams_stats_df = pd.concat([all_teams_stats_df, team_stats_df], ignore_index=True)
```

```python
all_teams_stats_df_1998_2022 = all_teams_stats_df[(all_teams_stats_df['YEAR'] >= '1998-99') & (all_teams_stats_df['YEAR'] <= '2022-23')]
```
